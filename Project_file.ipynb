## 6. Merge the two data frames so that all variables (columns) in the company's frame are added to the rounds2 data frame. Name the merged frame master_dataframe. How many observations are present in master_frame ?


```python
#To merge the dataframe we can either use .join() or .merge() function from pandas.
#We are providing information to python on how to merge the dataframes by passsing 'outer' in the how parameter.
#By mentioning right_index=True, left_index=True, we are asking python to join dataframes on the basis of both the index 
#right i.e d_c and left i,e d_r and join the dataframe.
#assigning the dataframe so obtained to a new variable called master_dataframe.
#master_dataframe=d_r.join(d_c,how='outer')....by using join 
master_dataframe = pd.merge(d_r, d_c, how="outer", right_index=True, left_index=True)
```

## 9. Create a pivot table to compare the mean and median values for ‘raised_amount_usd’ and ‘funding_round_type’ across different funding round categories..

## 11.Identify the top 9 countries in terms of highest investment in ‘venture’ fund type.


```python
#We can filter the data by using filtering method from pandas based on the conditions 
#We can pass condition in which we get data only where funding_round_type=='venture'.
#since we are looking for investment based on the countries, we are grouping data on 'country_code' column by pasiing it in 
#.groupby() function.
#And then getting total amount by using .sum() function. and sorting values in descending order by using .sort_values() fuction and passing ascending = False.
#By using .head() function we are getting only nine records from the data
#assigning the data to new variable called 'top_countries' and calling it.
top_countries=master_dataframe[master_dataframe.funding_round_type=='venture'].groupby(['country_code']).sum(['raised_amount_usd']).sort_values(by=['raised_amount_usd'],ascending=False).head(9)
top_countries
```

## 12. Create the ‘main_category’ column by extracting the main sector using the column‘category_list’.Hint - Use the Lambda function or string function.


```python
#Data in the 'main_Category' column has multiple values seperated by '|'.
#we can get only the first value by passing the data in .split() function. i.e  .split('|')[0]
#We can .split() on string datatypes.
# we can define this approach in a lambda function and then apply it on the data.
# assign the data so obtained from the lambda fucnction to a new column called 'main_Category'
#and call the dataframe.


main_category=[lambda x : x.split('|')[0] if type(x)==str else x]
master_dataframe['main_category']=master_dataframe.category_list.apply(main_category)
```

## 13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical column as follows
#### a) If the amount is less than 5 Millions then recode as TypeA.
#### b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
#### c) If the amount is greater than 7 Millions then recode as TypeC

```python
#To approach this function, first we have to define a user defined function, based on the conditions mentioned in the question.
#defined function can now be applied on the dataframe by using .apply() function.
#the data obtained is saved in a new variable called raised_aount_type and call it.
def raised_type (x):
    if x <= 5000000 :
        return 'TypeA'
    elif (x > 5000000) & (x < 7000000):
        return 'TypeB'
    elif x >= 7000000:
        return 'TypeC'
    else:
        return x
```

```python
raised_amount_type=pd.DataFrame(master_dataframe.raised_amount_usd.apply(raised_type))
```
